```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from pathlib import Path
from typing import Union
```

```python
hue_order = ["train", "val", "test"]
dx_order = ["akiec", "bcc", "bkl", "df", "mel", "nv", "vasc"]
```

```python
orig = pd.read_csv(Path("CSV_files/combined_metadata_original.csv"), header="infer")
# orig.groupby(["dx", "split"]).count()["image_id"]

orig_plot = sns.catplot(data=orig, kind="count", x="dx", hue="split", order=dx_order, hue_order=hue_order)
plt.yscale("log")
```

```python
corrected = pd.read_csv(Path("CSV_files/combined_metadata_corrected.csv"), header="infer")
# corrected.groupby(["dx", "split"]).count()["image_id"]

corrected_plot = sns.catplot(data=corrected, kind="count", x="dx", hue="split", order=dx_order, hue_order=hue_order)
plt.yscale("log")
```
